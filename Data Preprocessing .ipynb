### Define a mapping of similar city names to the desired common name

```python
import pandas as pd

# Assuming your DataFrame is named df
# Replace 'df' with the actual name of your DataFrame

# Define a dictionary to map similar city names to a common name
city_mapping = {
    '\\\\xc2\\\\xa0bangalore': 'Bangalore',
    '\\\\xc2\\\\xa0gurgaon': 'Gurgaon',
    '\\\\xc2\\\\xa0new delhi': 'New delhi',
    '\\\\xc2\\\\xa0mumbai': 'Mumbai',
    '\\\\xc2\\\\xa0Noida': 'Noida',
    '\\\\xc2\\\\xa0Bangalore': 'Bangalore',
    '\\\\xc2\\\\xa0Gurgaon': 'Gurgaon',
    '\\\\xc2\\\\xa0New Delhi': 'New delhi',
    '\\\\xc2\\\\xa0Mumbai': 'umbai',
    'Gurugram':'Gurgaon',
}

# Apply the city mapping to the 'city' column
df_Updated['City'] = df_Updated['City'].replace(city_mapping)
```
